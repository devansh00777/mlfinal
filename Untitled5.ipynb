```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from scipy.integrate import odeint
import torch
import torch.nn as nn

plt.style.use('seaborn-v0_8-darkgrid')
plt.rcParams.update({
    'font.family': 'DejaVu Serif',
    'font.size': 13,
    'axes.titlesize': 18,
    'axes.titleweight': 'bold',
    'axes.labelsize': 16,
    'axes.labelweight': 'bold',
    'legend.fontsize': 13,
    'xtick.labelsize': 13,
    'ytick.labelsize': 13,
    'figure.facecolor': 'white',
    'axes.facecolor': '#fafafa',
    'grid.alpha': 0.4,
    'grid.linestyle': '--',
    'lines.linewidth': 2.5,
})

alpha = 0.2
delta1 = 0.08
xi = 0.25
Lambda = 10.2
nu = 0.02
delta2 = 0.09
gamma = 0.01
mu = 9.0
beta_m = 1.5
beta_v = 0.57
beta_h = 0.765343138

# Initial conditions
S0 = 9.0
I0 = 0.1

# Time domain
t_span = np.linspace(0, 200, 2000)

# EXACT ODE system from paper (reduced SI model)
def ode_system(y, t):
    S, I = y
    dS = Lambda - (beta_h * S * I) / (1 + alpha * I) - (delta1 + nu) * S + (1 - beta_v) * delta2 * I
    dI = (beta_h * S * I) / (1 + alpha * I) + beta_v * delta2 * I - (delta2 + gamma) * I - (beta_m * I) / (1 + mu * I)
    return [dS, dI]

# Classical ODE solution
print(" Solving classical ODE...")
y0 = [S0, I0]
solution = odeint(ode_system, y0, t_span)
S_classical = solution[:, 0]
I_classical = solution[:, 1]

#: Add 5% Gaussian noise to I(t)
np.random.seed(42)
noise_level = 0.05
I_noisy = I_classical + noise_level * np.std(I_classical) * np.random.randn(len(I_classical))

t_tensor = torch.tensor(t_span.reshape(-1, 1), dtype=torch.float32, requires_grad=True)
I_data = torch.tensor(I_noisy.reshape(-1, 1), dtype=torch.float32)
S0_tensor = torch.tensor([[S0]], dtype=torch.float32)
I0_tensor = torch.tensor([[I0]], dtype=torch.float32)

# PINN Implementation
class PINN(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(1, 64),
            nn.Tanh(),
            nn.Linear(64, 64),
            nn.Tanh(),
            nn.Linear(64, 64),
            nn.Tanh(),
            nn.Linear(64, 64),
            nn.Tanh(),
            nn.Linear(64, 2)
        )

    def forward(self, t):
        out = self.net(t)
        S = torch.sigmoid(out[:, 0:1]) * 15.0
        I = torch.sigmoid(out[:, 1:2]) * 1000.0
        return S, I

model = PINN()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, factor=0.5, patience=500)

# Training
print("\n Training PINN")
epochs = 15000
t_colloc = t_tensor.clone().detach().requires_grad_(True)

loss_history = []
data_loss_history = []
physics_loss_history = []
ic_loss_history = []

for epoch in range(epochs):
    optimizer.zero_grad()

    # Predictions
    S_pred, I_pred = model(t_colloc)

    # Compute derivatives
    S_t = torch.autograd.grad(S_pred.sum(), t_colloc, create_graph=True)[0]
    I_t = torch.autograd.grad(I_pred.sum(), t_colloc, create_graph=True)[0]

    # Physics equations
    horiz_trans = (beta_h * S_pred * I_pred) / (1 + alpha * I_pred)
    treatment = (beta_m * I_pred) / (1 + mu * I_pred)
    vertical = beta_v * delta2 * I_pred

    # Residuals
    residual_S = S_t - (Lambda - horiz_trans - (delta1 + nu) * S_pred + (1 - beta_v) * delta2 * I_pred)
    residual_I = I_t - (horiz_trans + vertical - (delta2 + gamma) * I_pred - treatment)

    # Initial conditions
    t0 = torch.tensor([[0.0]], requires_grad=True)
    S0_pred, I0_pred = model(t0)

    # Data loss
    S_data_pred, I_data_pred = model(t_tensor)
    loss_data = torch.mean((I_data_pred - I_data)**2)

    # Physics loss
    loss_physics = torch.mean(residual_S**2) + torch.mean(residual_I**2)

    # Initial condition loss
    loss_ic = (S0_pred - S0_tensor)**2 + (I0_pred - I0_tensor)**2

    # Total loss
    loss = 1000.0 * loss_data + 10.0 * loss_physics + 500.0 * loss_ic

    loss.backward()
    optimizer.step()
    scheduler.step(loss)

    # Store losses
    loss_history.append(loss.item())
    data_loss_history.append(loss_data.item())
    physics_loss_history.append(loss_physics.item())
    ic_loss_history.append(loss_ic.item())

    if (epoch + 1) % 1500 == 0:
        print(f"Epoch {epoch+1:5d}/{epochs} | Loss: {loss.item():.4e} | Data: {loss_data.item():.4e} | Physics: {loss_physics.item():.4e}")

#  Predictions
print("\n Generating predictions...")
with torch.no_grad():
    S_pinn, I_pinn = model(t_tensor)
    S_pinn = S_pinn.numpy().flatten()
    I_pinn = I_pinn.numpy().flatten()

# Compute errors
rel_error_S = np.linalg.norm(S_pinn - S_classical) / np.linalg.norm(S_classical)
rel_error_I = np.linalg.norm(I_pinn - I_classical) / np.linalg.norm(I_classical)
rmse_S = np.sqrt(np.mean((S_pinn - S_classical)**2))
rmse_I = np.sqrt(np.mean((I_pinn - I_classical)**2))
mae_S = np.mean(np.abs(S_pinn - S_classical))
mae_I = np.mean(np.abs(I_pinn - I_classical))

print(f"\n{'='*70}")
print(" RESULTS:")
print(f"{'='*70}")
print(f"  Infected Population I(t):")
print(f"    â€¢ Relative L2 Error: {rel_error_I:.6f} ({rel_error_I*100:.3f}%)")
print(f"    â€¢ RMSE: {rmse_I:.6f}")
print(f"    â€¢ MAE:  {mae_I:.6f}")
print(f"\n  Susceptible Population S(t):")
print(f"    â€¢ Relative L2 Error: {rel_error_S:.6f} ({rel_error_S*100:.3f}%)")
print(f"    â€¢ RMSE: {rmse_S:.6f}")
print(f"    â€¢ MAE:  {mae_S:.6f}")
print(f"{'='*70}\n")

# Color palette (professional colors)
color_classical = '#1f77b4'  # Blue
color_pinn = '#d62728'       # Red
color_data = '#7f7f7f'       # Gray
color_error = '#ff7f0e'      # Orange
color_loss = '#9467bd'       # Purple

# ============================================================================
# PLOT 1: INFECTED POPULATION (Large, Standalone)
# ============================================================================
fig1 = plt.figure(figsize=(16, 9))
ax1 = fig1.add_subplot(111)

ax1.scatter(t_span[::5], I_noisy[::5], c=color_data, s=15, alpha=0.4,
            label='Noisy Training Data (5% noise)', zorder=1, marker='o', edgecolors='none')
ax1.plot(t_span, I_classical, color=color_classical, linewidth=4,
         label='Classical ODE (Ground Truth)', zorder=3, linestyle='-', alpha=0.9)
ax1.plot(t_span, I_pinn, color=color_pinn, linewidth=3.5,
         label='PINN Prediction', zorder=4, linestyle='--', alpha=0.95)

ax1.set_xlabel('Time (days)', fontweight='bold', fontsize=18)
ax1.set_ylabel('Infected Population I(t)', fontweight='bold', fontsize=18)
ax1.set_title('Infected Population: PINN vs Classical ODE Solution\n(Reduced SI Model with Saturated Incidence & Treatment)',
              fontsize=20, fontweight='bold', pad=25)
ax1.legend(loc='upper left', frameon=True, fancybox=True, shadow=True,
           fontsize=15, framealpha=0.95)
ax1.grid(True, alpha=0.35, linestyle='--', linewidth=1)
ax1.set_xlim([0, 200])

# Add detailed error box
textstr = (f'Performance Metrics:\n'
           f'â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”\n'
           f'Relative L2 Error: {rel_error_I:.6f}\n'
           f'RMSE: {rmse_I:.4f}\n'
           f'MAE: {mae_I:.4f}\n'
           f'Accuracy: {(1-rel_error_I)*100:.2f}%')
props = dict(boxstyle='round,pad=1', facecolor='#fff3cd', alpha=0.95,
             edgecolor='#856404', linewidth=3)
ax1.text(0.68, 0.35, textstr, transform=ax1.transAxes, fontsize=14,
         verticalalignment='top', bbox=props, family='monospace', weight='bold')

plt.tight_layout()
plt.savefig('01_infected_population.png', dpi=400, bbox_inches='tight', facecolor='white')
print("âœ… Saved: '01_infected_population.png'")
plt.show()

# ============================================================================
# PLOT 2: SUSCEPTIBLE POPULATION (Large, Standalone)
# ============================================================================
fig2 = plt.figure(figsize=(16, 9))
ax2 = fig2.add_subplot(111)

ax2.plot(t_span, S_classical, color=color_classical, linewidth=4,
         label='Classical ODE (Ground Truth)', zorder=2, linestyle='-', alpha=0.9)
ax2.plot(t_span, S_pinn, color=color_pinn, linewidth=3.5,
         label='PINN Prediction', zorder=3, linestyle='--', alpha=0.95)

ax2.set_xlabel('Time (days)', fontweight='bold', fontsize=18)
ax2.set_ylabel('Susceptible Population S(t)', fontweight='bold', fontsize=18)
ax2.set_title('Susceptible Population: PINN vs Classical ODE Solution\n(Reduced SI Model with Saturated Incidence & Treatment)',
              fontsize=20, fontweight='bold', pad=25)
ax2.legend(loc='best', frameon=True, fancybox=True, shadow=True,
           fontsize=15, framealpha=0.95)
ax2.grid(True, alpha=0.35, linestyle='--', linewidth=1)
ax2.set_xlim([0, 200])

# Add detailed error box
textstr = (f'Performance Metrics:\n'
           f'â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”\n'
           f'Relative L2 Error: {rel_error_S:.6f}\n'
           f'RMSE: {rmse_S:.4f}\n'
           f'MAE: {mae_S:.4f}\n'
           f'Accuracy: {(1-rel_error_S)*100:.2f}%')
props = dict(boxstyle='round,pad=1', facecolor='#d1ecf1', alpha=0.95,
             edgecolor='#0c5460', linewidth=3)
ax2.text(0.03, 0.97, textstr, transform=ax2.transAxes, fontsize=14,
         verticalalignment='top', bbox=props, family='monospace', weight='bold')

plt.tight_layout()
plt.savefig('02_susceptible_population.png', dpi=400, bbox_inches='tight', facecolor='white')
print("âœ… Saved: '02_susceptible_population.png'")
plt.show()

# ============================================================================
# PLOT 3: PHASE PORTRAIT (Large, Standalone)
# ============================================================================
fig3 = plt.figure(figsize=(14, 10))
ax3 = fig3.add_subplot(111)

ax3.plot(S_classical, I_classical, color=color_classical, linewidth=4,
         label='Classical ODE Trajectory', alpha=0.9)
ax3.plot(S_pinn, I_pinn, color=color_pinn, linewidth=3.5,
         label='PINN Trajectory', linestyle='--', alpha=0.9)
ax3.scatter([S0], [I0], s=400, c='green', marker='*',
            edgecolors='darkgreen', linewidths=3, zorder=5, label='Initial State (t=0)')
ax3.scatter([S_classical[-1]], [I_classical[-1]], s=300, c='navy',
            marker='s', edgecolors='black', linewidths=3, zorder=5, label='Final State (t=200)')

# Add arrows to show direction
arrow_indices = [200, 600, 1000, 1400]
for idx in arrow_indices:
    ax3.annotate('', xy=(S_classical[idx+50], I_classical[idx+50]),
                xytext=(S_classical[idx], I_classical[idx]),
                arrowprops=dict(arrowstyle='->', color=color_classical, lw=2.5, alpha=0.7))

ax3.set_xlabel('Susceptible Population S(t)', fontweight='bold', fontsize=18)
ax3.set_ylabel('Infected Population I(t)', fontweight='bold', fontsize=18)
ax3.set_title('Phase Portrait: System Trajectory in S-I Space\n(Reduced SI Model with Saturated Incidence & Treatment)',
              fontsize=20, fontweight='bold', pad=25)
ax3.legend(loc='upper right', frameon=True, fancybox=True, shadow=True, fontsize=14)
ax3.grid(True, alpha=0.35, linestyle='--', linewidth=1)

plt.tight_layout()
plt.savefig('03_phase_portrait.png', dpi=400, bbox_inches='tight', facecolor='white')
print("âœ… Saved: '03_phase_portrait.png'")
plt.show()

# ============================================================================
# PLOT 4: TRAINING CONVERGENCE (Large, Standalone)
# ============================================================================
fig4 = plt.figure(figsize=(16, 9))
ax4 = fig4.add_subplot(111)

epochs_plot = np.arange(1, len(loss_history) + 1)
ax4.semilogy(epochs_plot, loss_history, color=color_loss, linewidth=3.5,
             label='Total Loss', alpha=0.9)
ax4.semilogy(epochs_plot, data_loss_history, color=color_classical, linewidth=3,
             label='Data Loss', alpha=0.8, linestyle='-.')
ax4.semilogy(epochs_plot, physics_loss_history, color=color_pinn, linewidth=3,
             label='Physics Loss', alpha=0.8, linestyle=':')

ax4.set_xlabel('Training Epochs', fontweight='bold', fontsize=18)
ax4.set_ylabel('Loss (log scale)', fontweight='bold', fontsize=18)
ax4.set_title('PINN Training Convergence\n(15,000 Epochs with Adam Optimizer)',
              fontsize=20, fontweight='bold', pad=25)
ax4.legend(loc='upper right', frameon=True, fancybox=True, shadow=True, fontsize=15)
ax4.grid(True, alpha=0.35, linestyle='--', linewidth=1, which='both')

# Add final loss values
textstr = (f'Final Losses:\n'
           f'â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”\n'
           f'Total: {loss_history[-1]:.4e}\n'
           f'Data: {data_loss_history[-1]:.4e}\n'
           f'Physics: {physics_loss_history[-1]:.4e}')
props = dict(boxstyle='round,pad=1', facecolor='#e8daef', alpha=0.95,
             edgecolor='#7d3c98', linewidth=3)
ax4.text(0.72, 0.95, textstr, transform=ax4.transAxes, fontsize=14,
         verticalalignment='top', bbox=props, family='monospace', weight='bold')

plt.tight_layout()
plt.savefig('04_training_convergence.png', dpi=400, bbox_inches='tight', facecolor='white')
print("âœ… Saved: '04_training_convergence.png'")
plt.show()

# ============================================================================
# PLOT 5: PREDICTION ERROR (Large, Standalone)
# ============================================================================
fig5 = plt.figure(figsize=(16, 9))
ax5 = fig5.add_subplot(111)

error_I = I_pinn - I_classical
error_S = S_pinn - S_classical

ax5.plot(t_span, error_I, color=color_pinn, linewidth=3.5, label='Error in I(t)', alpha=0.9)
ax5.plot(t_span, error_S, color=color_error, linewidth=3.5, label='Error in S(t)', alpha=0.9)
ax5.axhline(y=0, color='black', linestyle='-', linewidth=2, alpha=0.7, label='Zero Error Line')
ax5.fill_between(t_span, 0, error_I, color=color_pinn, alpha=0.25)
ax5.fill_between(t_span, 0, error_S, color=color_error, alpha=0.25)

ax5.set_xlabel('Time (days)', fontweight='bold', fontsize=18)
ax5.set_ylabel('Prediction Error (PINN - ODE)', fontweight='bold', fontsize=18)
ax5.set_title('PINN Prediction Error Distribution Over Time\n(Deviation from Classical ODE Solution)',
              fontsize=20, fontweight='bold', pad=25)
ax5.legend(loc='best', frameon=True, fancybox=True, shadow=True, fontsize=15)
ax5.grid(True, alpha=0.35, linestyle='--', linewidth=1)
ax5.set_xlim([0, 200])

# Add error statistics
textstr = (f'Error Statistics:\n'
           f'â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”\n'
           f'I(t) - Max Error: {np.max(np.abs(error_I)):.4f}\n'
           f'I(t) - Mean Error: {np.mean(error_I):.4f}\n'
           f'S(t) - Max Error: {np.max(np.abs(error_S)):.4f}\n'
           f'S(t) - Mean Error: {np.mean(error_S):.4f}')
props = dict(boxstyle='round,pad=1', facecolor='#fadbd8', alpha=0.95,
             edgecolor='#cb4335', linewidth=3)
ax5.text(0.72, 0.97, textstr, transform=ax5.transAxes, fontsize=14,
         verticalalignment='top', bbox=props, family='monospace', weight='bold')

plt.tight_layout()
plt.savefig('05_prediction_error.png', dpi=400, bbox_inches='tight', facecolor='white')
print("âœ… Saved: '05_prediction_error.png'")
plt.show()

# ============================================================================
# PLOT 6: SIDE-BY-SIDE COMPARISON (Bonus)
# ============================================================================
fig6, axes = plt.subplots(1, 2, figsize=(20, 9))

# Infected
axes[0].scatter(t_span[::5], I_noisy[::5], c=color_data, s=12, alpha=0.4,
                label='Noisy Data (5%)', zorder=1)
axes[0].plot(t_span, I_classical, color=color_classical, linewidth=4,
             label='Classical ODE', zorder=2)
axes[0].plot(t_span, I_pinn, color=color_pinn, linewidth=3.5,
             label='PINN', linestyle='--', zorder=3, alpha=0.95)
axes[0].set_xlabel('Time (days)', fontweight='bold', fontsize=17)
axes[0].set_ylabel('Infected Population I(t)', fontweight='bold', fontsize=17)
axes[0].set_title('Infected Population Comparison', fontsize=19, fontweight='bold', pad=20)
axes[0].legend(fontsize=15, frameon=True, shadow=True, fancybox=True)
axes[0].grid(alpha=0.35, linestyle='--', linewidth=1)
textstr = f'Rel. L2 Error: {rel_error_I:.6f}\nRMSE: {rmse_I:.4f}\nAccuracy: {(1-rel_error_I)*100:.2f}%'
props = dict(boxstyle='round,pad=0.8', facecolor='wheat', alpha=0.9, edgecolor='#856404', linewidth=2.5)
axes[0].text(0.68, 0.32, textstr, transform=axes[0].transAxes, fontsize=14,
             bbox=props, family='monospace', weight='bold')

# Susceptible
axes[1].plot(t_span, S_classical, color=color_classical, linewidth=4,
             label='Classical ODE', zorder=2)
axes[1].plot(t_span, S_pinn, color=color_pinn, linewidth=3.5,
             label='PINN', linestyle='--', zorder=3, alpha=0.95)
axes[1].set_xlabel('Time (days)', fontweight='bold', fontsize=17)
axes[1].set_ylabel('Susceptible Population S(t)', fontweight='bold', fontsize=17)
axes[1].set_title('Susceptible Population Comparison', fontsize=19, fontweight='bold', pad=20)
axes[1].legend(fontsize=15, frameon=True, shadow=True, fancybox=True)
axes[1].grid(alpha=0.35, linestyle='--', linewidth=1)
textstr = f'Rel. L2 Error: {rel_error_S:.6f}\nRMSE: {rmse_S:.4f}\nAccuracy: {(1-rel_error_S)*100:.2f}%'
props = dict(boxstyle='round,pad=0.8', facecolor='lightblue', alpha=0.9, edgecolor='#0c5460', linewidth=2.5)
axes[1].text(0.03, 0.97, textstr, transform=axes[1].transAxes, fontsize=14,
             bbox=props, family='monospace', weight='bold')

plt.tight_layout()
plt.savefig('06_side_by_side_comparison.png', dpi=400, bbox_inches='tight', facecolor='white')
print("âœ… Saved: '06_side_by_side_comparison.png'")
plt.show()

print("\n" + "="*70)
print("ðŸŽ‰ ALL DONE! 6 Beautiful Separate Plots Generated Successfully!")
print("="*70)
print("\nGenerated Files:")
print("  1. 01_infected_population.png")
print("  2. 02_susceptible_population.png")
print("  3. 03_phase_portrait.png")
print("  4. 04_training_convergence.png")
print("  5. 05_prediction_error.png")
print("  6. 06_side_by_side_comparison.png")
print("="*70)
```
